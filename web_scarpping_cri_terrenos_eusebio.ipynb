```python

from requests_html import HTMLSession
from bs4 import BeautifulSoup
import pandas as pd
import csv
from time import sleep
from random import randint
import numpy as np
import requests
```

```python

url = 'https://www.cri-ce.com.br/imoveis/a-venda/terreno/eusebio'
session = HTMLSession()
r1=session.get(url)

sp = BeautifulSoup(r1.content, 'html5lib')
tot_anuncios = int(sp.find("span", class_="h-money").text)

anuncios_por_pagina = 12

if tot_anuncios % anuncios_por_pagina == 0:
    pag = tot_anuncios/anuncios_por_pagina
    print(pag)
elif tot_anuncios % anuncios_por_pagina != 0:
    pag = int(tot_anuncios/anuncios_por_pagina) + 1
    print(pag)


```

```python

headers = dict()
headers[
    "User-Agent"
] = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36"

descricoes = []
tipos = []
locais = []
valores = []
areas = []

pages = np.arange(1, 3) #N. de pag +1

for page in pages:
  if page>=2:url = "https://www.cri-ce.com.br/imoveis/a-venda/terreno/eusebio?pagina="+ str(page) 
  else:url = "https://www.cri-ce.com.br/imoveis/a-venda/terreno/eusebio"
      
  print(url)
  results = requests.get(url, headers=headers)
  soup = BeautifulSoup(results.text, "html.parser")

  imovel_div = soup.find_all("div", class_="link-all")
  #print(imovel_div)
  sleep(randint(2, 10))
  #print(page)
  for imovelSection in imovel_div:
      
      localSection = imovelSection.find('div',class_='card-block')
      local = localSection.find('h2',class_='card-title').text
      locais.append(local)

      tipo = localSection.find('h3',class_='card-text').text
      tipos.append(tipo)

      #descricao = localSection.find('p',class_='description hidden-sm-down').text
      #descricoes.append(descricao)
      #print(descricao)

      valor = imovelSection.find("span", class_='h-money location').text.strip()
      valores.append(valor)
      
      areaSection=imovelSection.find('div',class_='values')
      area = areaSection.find("span", class_='h-money').text.strip()
      areas.append(area)
      #print(tipo)
      
eusebio = pd.DataFrame(
    {
        "Locais": locais,
        "Tipos": tipos,
        "Valores": valores,
        "Areas": areas
    }
)
eusebio
```
